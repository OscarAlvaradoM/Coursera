```julia
# Package to read .dat document
using DelimitedFiles
# Reading data
data_tofit = readdlm("Week3_PR_Data.dat")
```

```julia
# The variables we have to plot
x = data_tofit[:,1]
y = data_tofit[:,2];
```

```julia
# Plot the x and y data points using a scatter plot of the x and y array variables
using Plots
gr() # Activate the GR backend for use with Plots
scatter(x,y)
```

```julia
# Create a function called parabfit, with x as the argument, returning a*x^2 + b*x + c
parabfit(x) = a*x^2 + b*x + c
```

```julia
# Create variables a, b and c, assigning each the value 1
a = b = c = 1
plot(parabfit, -5,5)
```

```julia
a = -.95
b = 3.5
c = -1.5
scatter(x,y, label = "Data", c = :black)
plot!(parabfit,0,6,label = "Model", c = :orange, width = 2)
plot!(title = "Fitting a curve")
```

Initially _a_, _b_ and _c_ were not very good, first of all, _a_ was positive, so it was the clearest thing to change, then it was clear that _c_ has to be negative too because near zero, data seems to be negative. At the end I just change b until the parabola fit well enough.
